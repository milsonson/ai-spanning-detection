```python
def count_calls(m):
    if m == 1:
        return 1
    elif m == 2:
        return 1
    else:
        return count_calls(m - 1) + count_calls(m - 2) + 1
import time
from tqdm import tqdm

def decorator(func):
    call_count = 0
    total_calls = 0

    def wrapper(*args, **kwargs):
        nonlocal call_count, total_calls

        if call_count == 0:
            total_calls = count_calls(*args, **kwargs)
            pbar = tqdm(total=total_calls, desc="Calculating my_func")

        call_count += 1
        pbar.update(1)

        a = time.time()
        ans = func(*args, **kwargs)
        a = time.time() - a

        if call_count == total_calls:
            pbar.close()

        print(f"Time taken: {a} seconds")
        return ans

    def count_calls(m):
        if m == 1:
            return 1
        elif m == 2:
            return 1
        else:
            return count_calls(m - 1) + count_calls(m - 2) + 1

    return wrapper

@decorator
def my_func(m):
    if m == 1:
        return 1
    elif m == 2:
        return 2
    else:
        return my_func(m - 1) + my_func(m - 2)

print(my_func(10))
```
